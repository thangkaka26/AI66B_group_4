```python
# IMPORTS

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mlp_loss = pd.DataFrame()
mlp_loss['RMSE_LSTM'] = [1218.6, 1659.6, 2005.2]

mlp_loss['RMSE_FFNN'] = [3203.6, 3745.6, 4877.8]

mlp_loss['MAE_LSTM'] = [1007.4, 1291.4, 1733.8]

mlp_loss['MAE_FFNN'] = [2601.2, 2965.2, 4007.1]

mlp_loss.index = ["ZONE 1", "ZONE 2", "ZONE 3"]
display(mlp_loss)


x = np.arange(3)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,8))
fig.suptitle("LSTM & FFNN Comparisons", fontsize=22, fontweight="bold")

ax1.bar(x-0.15, mlp_loss['RMSE_LSTM'], 0.3, label='LSTM (Ours)')
ax1.bar(x+0.2, mlp_loss['RMSE_FFNN'], 0.3, label='FFNN (Paper)')
ax1.set_xticks(x, mlp_loss.index, fontsize=14)
ax1.legend(fontsize=14)
ax1.set_facecolor('#F5FFFA')
ax1.grid(axis="y", linestyle="--", alpha=0.5)
ax1.set_title("RMSE", fontsize=20, fontweight="bold")
ax1.set_ylabel("Power (W)", fontsize=14)
for container in ax1.containers:
      ax1.bar_label(container, fontsize=12)

ax2.bar(x-0.15, mlp_loss['MAE_LSTM'], 0.3, label='LSTM (Ours)')
ax2.bar(x+0.2, mlp_loss['MAE_FFNN'], 0.3, label='FFNN (Paper)')
ax2.set_xticks(x, mlp_loss.index, fontsize=14)
ax2.grid(axis="y", linestyle="--", alpha=0.5)
ax2.legend(fontsize=14)
ax2.set_facecolor('#F5FFFA')
ax2.set_title("MAE", fontsize=20, fontweight="bold")
ax2.set_ylabel("Power (W)", fontsize=14)
for container in ax2.containers:
      ax2.bar_label(container, fontsize=12)

plt.show()
```
